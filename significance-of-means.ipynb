## Significance of means 
Determining if the difference in means in thw-only and thw-or-mhw are statistically different at each grid point, using Scipy T test 

```python
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import ttest_ind 
import xarray as xr
```

```python
thw_only = xr.open_dataset('/g/data/e14/cp3790/Charuni/Heatwaves/thw-only-2.nc')
```

```python
thw_or_mhw = xr.open_dataset('/g/data/e14/cp3790/Charuni/Heatwaves/thw-or-mhw-2.nc')
```

```python
fig, ax = plt.subplots(figsize=(8,6)) 
ploteo=thw_only.event_duration_x.mean(dim='event').plot(
    cmap=plt.cm.get_cmap('plasma', 6),  
    vmin=3, vmax=9, # Minimum and maximum values of the colormap
)
ploteo.colorbar.ax.set_ylabel('Event duration',fontsize=12)
plt.xlabel('Longitude',fontsize=16)
plt.ylabel('Latitude', fontsize=16)
plt.title('Mean duration of stand alone THW events', fontsize=14)
#plt.savefig('Duration-stand-alone-events-AUS')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6)) 
ploteo=thw_or_mhw.event_duration_x.mean(dim='event').plot(
    cmap=plt.cm.get_cmap('plasma', 17),  
    vmin=3, vmax=20, # Minimum and maximum values of the colormap
)
ploteo.colorbar.ax.set_ylabel('Event duration',fontsize=12)
plt.xlabel('Longitude',fontsize=16)
plt.ylabel('Latitude', fontsize=16)
plt.title('Mean duration of stand alone THW events', fontsize=14)
#plt.savefig('Duration-stand-alone-events-AUS')
plt.show()
```

```python
r =stats.ttest_ind(thw_only.event_duration_x, thw_or_mhw.event_duration_x, axis=0, equal_var=False)
```
